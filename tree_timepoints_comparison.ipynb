Plot to generate the comparison of the different errors 
Uses data taken from the hyperparameter tuning from ibtracts_sst_rf and ibtracs_sst_xgboost

```python
import matplotlib.pyplot as plt

t = [1,3,5,10,15]
xgb_ws_err = [4.82880168,3.74388301,3.82653792,3.91722686,3.98680861]
rf_ws_err = [5.35332694,4.67110725,4.75908144,4.85333398,4.94124005]

plt.figure(figsize=(10,4))
plt.plot(t,xgb_ws_err,label="XGBoost")
plt.plot(t,rf_ws_err,label="RF")
plt.xlabel("Number of prior observations")
plt.ylabel("RMSE Error in wind speed (kts) ")
plt.legend()
plt.savefig("model_comp.png")
```
