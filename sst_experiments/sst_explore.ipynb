### explore and validate SST interpolation data

```python
import xarray
ds = xarray.open_dataset('../data/sst.wkmean.1990-present.nc')
```

```python
ds = ds.assign_coords(lon=(((ds.lon + 180) % 360) - 180))
```

```python
from datetime import datetime
ds.sst.interp(time='2000-01-01',lat=0.0,lon=0.0,method='linear')
```
